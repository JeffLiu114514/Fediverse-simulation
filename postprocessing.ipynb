```python
import json
import pandas as pd
import matplotlib.pyplot as plt

n_nodes_list = [x for x in range(50, 501, 25)]
n_posts_list = [x for x in range(50, 501, 25)]
variances = [x/100 for x in range(10, 101, 10)]

data = json.load(open("results.json", "r"))

```

```python
# n_nodes, n_posts, variance
def plot_results(data, plot_var, fixed_var1, fixed_var2):
    kdt_distance_normalized = []
    if plot_var == "n_nodes":
        plot_var_list = n_nodes_list
        for n_nodes in n_nodes_list:
            kdt_distance_normalized.append(data[str((n_nodes, fixed_var1, fixed_var2))]["Average_Normalized_Distance"])
    elif plot_var == "n_posts":
        plot_var_list = n_posts_list
        for n_posts in n_posts_list:
            kdt_distance_normalized.append(data[str((fixed_var1, n_posts, fixed_var2))]["Average_Normalized_Distance"])
    elif plot_var == "variance":
        plot_var_list = variances
        for variance in variances:
            kdt_distance_normalized.append(data[str((fixed_var1, fixed_var2, variance))]["Average_Normalized_Distance"])
    
    df = pd.DataFrame({plot_var: plot_var_list, 'Kendall-Tau-Distance-Normalized': kdt_distance_normalized})
    df.plot(x=plot_var, y='Kendall-Tau-Distance-Normalized')
```

```python
plot_results(data, "n_nodes", 100, 0.1)
```
